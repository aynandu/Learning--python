```python
#1 - Import Data
import pandas as pd
dataFrame=pd.read_csv('./stat/FIFA23_official_data.csv')
dataFrame.shape
```

```python
#2 - Clean the Data
df=pd.DataFrame(dataFrame,columns=['Name','Wage','Value'])
def ValueToFloat(x):
    if type(x)==float or type(x)==int:
        return x
    if 'K' in x:
        if len(x)>1:
            return float (x.replace('K',''))*1000
        return 1000.0
    if 'M' in x:
        if len(x)>1:
            return float (x.replace('M',''))*1000000
        return 1000000.0
    if 'B' in x:
        if len(x)>1:
            return float (x.replace('B',''))*1000000000
        return 0.0  
wage=df['Wage'].replace('[\€,]','',regex=True).apply(ValueToFloat)
value=df['Value'].replace('[\€,]','',regex=True).apply(ValueToFloat)
df['Wage']=wage
df['Value']=value
df['difference']=df['Value']-df['Wage']
df.sort_values('difference',ascending=False)
```

```python
import seaborn as sns
sns.set()
graph=sns.scatterplot(x='Wage',y='Value',data=df)
graph
```

```python
#2 - Clean the Data
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
TOOLTIPS=HoverTool(tooltips=[
    ("index","$index"),
    ("(Wage,Value)","(@Wage,@Value)"),
    ("Name","@Name")
])
p=figure(title="Soccer 2023",x_axis_label='Wage',y_axis_label='Value',width=700,height=700,tools=[TOOLTIPS])
p.star('Wage','Value',size=10, source=df)
#p = figure(x_axis_label='Wage',y_axis_label='Value', height=350, title="Fruit Counts",toolbar_location=None, tools="")
show(p)

#https://docs.bokeh.org/en/latest/docs/user_guide/basic/scatters.html#ug-basic-scatters
```
